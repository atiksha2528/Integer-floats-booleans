```python
#usinf for loop
even_squared_numbers=[]
for number in list_numbers:
    if number%2==0:
        even_squared_numbers.append(number**2)
print(f"even_squared_numbers: {even_squared_numbers}")
```

```python
#using list comprehension
even_squared_numbers=[number**2 for number in list_numbers if number%2==0]
print(f"even_squared_numbers: {even_squared_numbers}")
```

```python
#dictionary comprehension
country_city_dict={"china":"beijing","india":"new delhi","usa":"washington dc"}
print(country_city_dict)
```

```python
#swap the keys and values for this dictionary
flipped={}
for country,city in country_city_dict.items():
    flipped[city]=country
print(flipped)
```

```python
flipped_dict={city:country for country,city in country_city_dict.items()}
print(flipped_dict)
```

```python
flipped={country[:2]:city for country,city in country_city_dict.items()}
print(flipped)
```

```python
flipped={country[:2].upper():city.upper() for country,city in country_city_dict.items()}
print(flipped)
```
